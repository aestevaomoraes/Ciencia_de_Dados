## Distribui√ß√µes Estat√≠sticas: ‚úÖ 7.7 Regra Emp√≠rica (68-95-99.7)

### üìå Introdu√ß√£o

A Regra Emp√≠rica (ou Regra 68-95-99.7) √© um princ√≠pio fundamental na estat√≠stica que descreve como os dados s√£o distribu√≠dos em uma distribui√ß√£o normal. Essa regra afirma que:

* Aproximadamente **68%** dos dados est√£o dentro de `1 desvio padr√£o da m√©dia.`

* Aproximadamente **95%** dos dados est√£o dentro de `2 desvios padr√£o da m√©dia.`

* Aproximadamente **99.7%** dos dados est√£o dentro de `3 desvios padr√£o da m√©dia.`

Essa regra √© amplamente utilizada para an√°lise estat√≠stica, detec√ß√£o de outliers e previs√µes em diversos campos, como economia, engenharia e ci√™ncias sociais

### üìå Exemplo Pr√°tico do Dia a Dia

Imagine que a altura dos adultos em uma cidade segue uma distribui√ß√£o normal com m√©dia de 170 cm e desvio padr√£o de 10 cm. Aplicando a Regra Emp√≠rica:

* **68%** das pessoas t√™m altura entre `160 cm e 180 cm.`

* **95%**  das pessoas t√™m altura entre `150 cm e 190 cm.`

* **99.7%** das pessoas t√™m altura entre `140 cm e 200 cm.`

### üìå Implementa√ß√£o em Python

Podemos visualizar essa distribui√ß√£o e validar a Regra Emp√≠rica com Python usando **Jupyter Notebook** ou **VS Code**.

#### üìç Criando um Dataset Fict√≠cio

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Gerando dados fict√≠cios seguindo uma distribui√ß√£o normal
np.random.seed(42)
data = np.random.normal(loc=170, scale=10, size=1000)

# Criando um DataFrame
df = pd.DataFrame(data, columns=['Altura'])

# Visualizando as primeiras linhas
df.head()
```

### üìç Visualiza√ß√£o Gr√°fica

```python
# Criando um histograma com curva de densidade
plt.figure(figsize=(10, 6))
sns.histplot(df['Altura'], kde=True, bins=30, color='blue')
plt.axvline(170, color='red', linestyle='dashed', label='M√©dia')
plt.axvline(160, color='green', linestyle='dashed', label='1 Desvio Padr√£o')
plt.axvline(180, color='green', linestyle='dashed')
plt.axvline(150, color='orange', linestyle='dashed', label='2 Desvios Padr√£o')
plt.axvline(190, color='orange', linestyle='dashed')
plt.axvline(140, color='purple', linestyle='dashed', label='3 Desvios Padr√£o')
plt.axvline(200, color='purple', linestyle='dashed')
plt.legend()
plt.title('Distribui√ß√£o Normal das Alturas')
plt.show()
```
