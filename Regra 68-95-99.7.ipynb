## Distribuições Estatísticas: ✅ 7.7 Regra Empírica (68-95-99.7)

### 📌 Introdução

A Regra Empírica (ou Regra 68-95-99.7) é um princípio fundamental na estatística que descreve como os dados são distribuídos em uma distribuição normal. Essa regra afirma que:

* Aproximadamente **68%** dos dados estão dentro de `1 desvio padrão da média.`

* Aproximadamente **95%** dos dados estão dentro de `2 desvios padrão da média.`

* Aproximadamente **99.7%** dos dados estão dentro de `3 desvios padrão da média.`

Essa regra é amplamente utilizada para análise estatística, detecção de outliers e previsões em diversos campos, como economia, engenharia e ciências sociais

### 📌 Exemplo Prático do Dia a Dia

Imagine que a altura dos adultos em uma cidade segue uma distribuição normal com média de 170 cm e desvio padrão de 10 cm. Aplicando a Regra Empírica:

* **68%** das pessoas têm altura entre `160 cm e 180 cm.`

* **95%**  das pessoas têm altura entre `150 cm e 190 cm.`

* **99.7%** das pessoas têm altura entre `140 cm e 200 cm.`

### 📌 Implementação em Python

Podemos visualizar essa distribuição e validar a Regra Empírica com Python usando **Jupyter Notebook** ou **VS Code**.

#### 📍 Criando um Dataset Fictício

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Gerando dados fictícios seguindo uma distribuição normal
np.random.seed(42)
data = np.random.normal(loc=170, scale=10, size=1000)

# Criando um DataFrame
df = pd.DataFrame(data, columns=['Altura'])

# Visualizando as primeiras linhas
df.head()
```

### 📍 Visualização Gráfica

```python
# Criando um histograma com curva de densidade
plt.figure(figsize=(10, 6))
sns.histplot(df['Altura'], kde=True, bins=30, color='blue')
plt.axvline(170, color='red', linestyle='dashed', label='Média')
plt.axvline(160, color='green', linestyle='dashed', label='1 Desvio Padrão')
plt.axvline(180, color='green', linestyle='dashed')
plt.axvline(150, color='orange', linestyle='dashed', label='2 Desvios Padrão')
plt.axvline(190, color='orange', linestyle='dashed')
plt.axvline(140, color='purple', linestyle='dashed', label='3 Desvios Padrão')
plt.axvline(200, color='purple', linestyle='dashed')
plt.legend()
plt.title('Distribuição Normal das Alturas')
plt.show()
```

### 📌 Conclusão

A Regra Empírica fornece uma maneira intuitiva de compreender a dispersão dos dados em uma distribuição normal. 

Aplicada corretamente, pode auxiliar na interpretação de fenômenos naturais, na análise de qualidade e no controle estatístico de processos.
