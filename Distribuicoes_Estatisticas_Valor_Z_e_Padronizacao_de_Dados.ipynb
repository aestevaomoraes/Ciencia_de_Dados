## üìå Introdu√ß√£o

O **Valor Z** (ou **escore Z**) √© uma medida estat√≠stica que indica quantos desvios padr√£o um valor est√° distante da m√©dia de uma distribui√ß√£o. 
Ele √© amplamente utilizado para padronizar dados, comparar diferentes distribui√ß√µes e identificar outliers. 

### üìå Conceitos Acad√™micos

#### 7.9.1 O que √© o Valor Z?
O Valor Z √© uma medida de quantos desvios padr√£o um valor est√° acima ou abaixo da m√©dia de uma distribui√ß√£o. Ele √© calculado pela f√≥rmula:



#### 7.9.2 F√≥rmula do Valor Z
A f√≥rmula do Valor Z √© simples, mas poderosa. Ela transforma qualquer valor em uma escala padronizada, onde:

* **Z=0:** O valor √© igual √† m√©dia.

* **Z>0:** O valor est√° acima da m√©dia.

* **Z<0:** O valor est√° abaixo da m√©dia.

#### 7.9.3 Aplica√ß√µes do Valor Z
O Valor Z √© usado em diversas √°reas, como:

* **Padroniza√ß√£o de dados:** Transformar diferentes escalas em uma escala comum.

* **Testes de hip√≥tese:** Comparar m√©dias de diferentes amostras.

* **Identifica√ß√£o de outliers:** Valores com `‚à£Z‚à£>3` s√£o considerados outliers.

### üíª Pr√°tica: Implementa√ß√£o em Python
Vamos implementar o conceito de Valor Z em Python, usando bibliotecas como NumPy, Pandas, Matplotlib e Seaborn.

#### 1. Gerando um Dataset Fict√≠cio
Primeiro, vamos criar um dataset fict√≠cio com uma distribui√ß√£o normal.

```python
import numpy as np
import pandas as pd

# Gerando dados fict√≠cios
np.random.seed(42)
dados = np.random.normal(loc=50, scale=10, size=1000)  # M√©dia = 50, Desvio = 10

# Criando um DataFrame
df = pd.DataFrame(dados, columns=['Valores'])
df.head()
```

### 2. Calculando o Valor Z
Agora, vamos calcular o Valor Z para cada valor no dataset.

```python
# Importando bibliotecas necess√°rias
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Gerando um dataset fict√≠cio com distribui√ß√£o normal
np.random.seed(42)  # Define uma semente para reproducibilidade
dados = np.random.normal(loc=50, scale=10, size=1000)  # M√©dia = 50, Desvio Padr√£o = 10

# Criando um DataFrame com os dados gerados
df = pd.DataFrame(dados, columns=['Valores'])

# Calculando a m√©dia e o desvio padr√£o do dataset
media = df['Valores'].mean()  # M√©dia dos valores
desvio_padrao = df['Valores'].std()  # Desvio padr√£o dos valores

# Exibindo a m√©dia e o desvio padr√£o calculados
print(f"M√©dia: {media:.2f}")
print(f"Desvio Padr√£o: {desvio_padrao:.2f}")

# Plotando a distribui√ß√£o dos dados usando Seaborn
sns.histplot(df['Valores'], kde=True, color='skyblue', label='Distribui√ß√£o dos Dados')

# Adicionando linhas verticais para representar a m√©dia e os desvios padr√£o
plt.axvline(media, color='red', linestyle='--', label=f'M√©dia: {media:.2f}')  # Linha da m√©dia
plt.axvline(media + desvio_padrao, color='green', linestyle=':', label='¬±1 Desvio Padr√£o')  # +1 desvio padr√£o
plt.axvline(media - desvio_padrao, color='green', linestyle=':')  # -1 desvio padr√£o
plt.axvline(media + 2*desvio_padrao, color='orange', linestyle='-.', label='¬±2 Desvios Padr√£o')  # +2 desvios padr√£o
plt.axvline(media - 2*desvio_padrao, color='orange', linestyle='-.')  # -2 desvios padr√£o
plt.axvline(media + 3*desvio_padrao, color='purple', linestyle='--', label='¬±3 Desvios Padr√£o')  # +3 desvios padr√£o
plt.axvline(media - 3*desvio_padrao, color='purple', linestyle='--')  # -3 desvios padr√£o

# Adicionando t√≠tulo e r√≥tulos aos eixos
plt.title('Distribui√ß√£o Normal com Valor Z')
plt.xlabel('Valores')
plt.ylabel('Frequ√™ncia')

# Adicionando uma legenda ao gr√°fico
plt.legend()

# Exibindo o gr√°fico
plt.show()
```

### 4. Identificando Outliers com o Valor Z
Podemos usar o Valor Z para identificar outliers no dataset.

```python
# Importando bibliotecas necess√°rias
import numpy as np
import pandas as pd

# Gerando um dataset fict√≠cio com distribui√ß√£o normal
np.random.seed(42)  # Define uma semente para reproducibilidade
dados = np.random.normal(loc=50, scale=10, size=1000)  # M√©dia = 50, Desvio Padr√£o = 10

# Criando um DataFrame com os dados gerados
df = pd.DataFrame(dados, columns=['Valores'])

# Calculando a m√©dia e o desvio padr√£o do dataset
media = df['Valores'].mean()  # M√©dia dos valores
desvio_padrao = df['Valores'].std()  # Desvio padr√£o dos valores

# Calculando o Valor Z para cada valor no dataset
df['Z'] = (df['Valores'] - media) / desvio_padrao

# Exibindo as primeiras linhas do DataFrame com os valores Z calculados
print("DataFrame com Valores Z calculados:")
print(df.head())

# Identificando outliers (|Z| > 3)
outliers = df[np.abs(df['Z']) > 3]

# Exibindo os outliers identificados
print("\nOutliers identificados (|Z| > 3):")
print(outliers)
```
