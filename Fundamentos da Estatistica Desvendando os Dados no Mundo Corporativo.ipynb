### 1.4 Popula√ß√£o e Amostra üåç

* **Popula√ß√£o:** Todo o grupo de interesse (ex.: todos os funcion√°rios de uma empresa).
* **Amostra:** Uma parte representativa (ex.: 50 funcion√°rios).
  
 

### 1.5 Vari√°veis: Qualitativas vs. Quantitativas üìà
* **Qualitativas:** Categorias.
  * Nominais: Sem ordem (ex.: departamento ‚Äì TI, Vendas).
  * Ordinais: Com ordem (ex.: satisfa√ß√£o ‚Äì baixa, m√©dia, alta).
* **Quantitativas:** N√∫meros.
  * Discretas: Cont√°veis (ex.: n√∫mero de vendas).
  * Cont√≠nuas: Intervalos (ex.: horas de trabalho).



### Exemplo Pr√°tico com Python üêç

Imagine que voc√™ √© um gestor analisando **"Horas de Treinamento por Semana"**. Vamos criar um dataset fict√≠cio e visualiz√°-lo!

```python
# Importando bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Configurando fonte para suportar emojis (opcional)
plt.rcParams['font.family'] = 'Segoe UI Emoji'

# Dataset fict√≠cio
dados = {
    "Colaborador": ["Ana", "Bruno", "Carla", "Diego", "Eva"],
    "Horas_Treinamento": [2.5, 1.0, 4.0, 1.5, 3.5],  # Quantitativa Cont√≠nua
    "Departamento": ["TI", "Vendas", "RH", "TI", "RH"]  # Qualitativa Nominal
}
df = pd.DataFrame(dados)

# Tabela de frequ√™ncia
print("üìä Frequ√™ncia por Departamento:")
print(df["Departamento"].value_counts())

# Gr√°fico de barras
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Departamento", hue="Departamento", palette="Blues", legend=False)
plt.title("Distribui√ß√£o por Departamento üìà")
plt.xlabel("Departamento")
plt.ylabel("Frequ√™ncia")
plt.show()

# Histograma com anota√ß√µes e √°rea sombreada
plt.figure(figsize=(6, 4))
sns.histplot(df["Horas_Treinamento"], bins=4, kde=True, color="blue", alpha=0.5)
plt.title("Horas de Treinamento por Semana üìä")
plt.xlabel("Horas")
plt.ylabel("Frequ√™ncia")

# Linha da m√©dia
media = df["Horas_Treinamento"].mean()
plt.axvline(media, color="red", linestyle="--", label=f"M√©dia: {media:.2f}h")

# √Årea sombreada (ex.: abaixo da m√©dia)
x = np.linspace(0, media, 100)
y = sns.kdeplot(df["Horas_Treinamento"]).get_lines()[0].get_data()[1]
plt.fill_between(x, 0, np.interp(x, sns.kdeplot(df["Horas_Treinamento"]).get_lines()[0].get_data()[0], y), 
                 color="red", alpha=0.3, label="Abaixo da m√©dia")

# Anota√ß√£o
plt.text(media + 0.2, plt.ylim()[1] * 0.8, f"M√©dia: {media:.2f}h", color="red")
plt.legend()
plt.show()
```

### Sa√≠da: 
* Tabela: TI (2), RH (2), Vendas (1).
* Gr√°fico de barras: Mostra departamentos mais treinados.
* **Histograma:** Distribui√ß√£o das horas, com m√©dia em vermelho.
  
### Interpreta√ß√£o Simples:
* TI e RH lideram em treinamentos.
* A maioria gasta entre 1 e 4 horas, com m√©dia de ~2.5h.
* Esse insight pode guiar aloca√ß√£o de recursos no RH!
