```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
import matplotlib.font_manager as fm

# Remover a definição de fonte para emojis, usando a configuração padrão
plt.rcParams['font.family'] = 'sans-serif'

# Introdução
print("Semana 3 – Distribuições Estatísticas: Valor Z e Padronização de Dados")
print("7.9.1 O que é o Valor Z?")
print("O valor Z mede a posição de um dado em relação à média de uma distribuição normal, em termos de desvio-padrão.")

# 7.9.2 Fórmula do Valor Z
print("7.9.2 Fórmula do Valor Z")
print("Fórmula: Z = (X - μ) / σ")
print("Onde:")
print("- X é o valor da amostra")
print("- μ é a média da população")
print("- σ é o desvio-padrão")

# 7.9.3 Aplicações do Valor Z
print("7.9.3 Aplicações do Valor Z")
print("O valor Z é usado para determinar a probabilidade de um evento ocorrer dentro de uma distribuição normal.")

# Simulação de uma Distribuição Normal
dados = np.random.normal(loc=100, scale=15, size=1000)  # Média 100, Desvio padrão 15

# Cálculo do Valor Z
def calcular_valor_z(x, media, desvio):
    return (x - media) / desvio

media_amostral = np.mean(dados)
desvio_amostral = np.std(dados, ddof=1)
dados_z = np.array([calcular_valor_z(x, media_amostral, desvio_amostral) for x in dados])

# Criar um DataFrame com os dados
df = pd.DataFrame({'Valores Originais': dados, 'Valor Z': dados_z})

# Visualização da Distribuição Original
plt.figure(figsize=(10, 5))
sns.histplot(dados, kde=True, bins=30, color='blue')
plt.axvline(media_amostral, color='red', linestyle='dashed', linewidth=2, label='Média')
plt.title("Distribuição Normal Simulada")
plt.legend()
plt.show()

# Visualização da Distribuição dos Valores Z
plt.figure(figsize=(10, 5))
sns.histplot(dados_z, kde=True, bins=30, color='green')
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Média (Z=0)')
plt.title("Distribuição dos Valores Z")
plt.legend()
plt.show()

# Conclusão
print("Conclusão")
print("O valor Z nos ajuda a entender como um dado se compara à média da distribuição e permite calcular probabilidades estatísticas.")

```
