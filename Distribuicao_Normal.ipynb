## Distribuição Normal: Entenda e Visualize seus Dados com Python

A distribuição normal, também conhecida como distribuição gaussiana, é uma das mais importantes distribuições de probabilidade em estatística. Sua forma de sino simétrica e suas propriedades únicas a tornam fundamental para modelar uma ampla gama de fenômenos do mundo real.

### Por que a distribuição normal é tão importante?

* **Teorema do Limite Central:** Este teorema fundamental afirma que a soma (ou média) de um grande número de variáveis aleatórias independentes e identicamente distribuídas (independentemente de sua distribuição original) se aproxima de uma distribuição normal. Isso explica por que a distribuição normal aparece com tanta frequência em diversas áreas.

* **Inferência Estatística:**  Muitos testes estatísticos e modelos de machine learning assumem que os dados seguem uma distribuição normal. Isso permite fazer inferências e previsões confiáveis.

* **Simplicidade e Modelagem:** A distribuição normal é definida por apenas dois parâmetros: média (μ) e desvio padrão (σ). Isso a torna fácil de entender e usar para modelar dados.

### Conceitos Estatísticos Resumidos:

* **Média (μ):** O valor central da distribuição.   
* **Desvio Padrão (σ):**  Mede a dispersão dos dados em torno da média.   
* **68-95-99.7 Regra:** Aproximadamente 68% dos dados estão dentro de 1 desvio padrão da média, 95% dentro de 2 desvios padrão e 99.7% dentro de 3 desvios padrão.

Criando a Série de Dados de Exemplo com Python:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Definindo os parâmetros da distribuição normal
media = 0
desvio_padrao = 1
tamanho_da_amostra = 1000

# Gerando dados aleatórios com distribuição normal
dados = np.random.normal(media, desvio_padrao, tamanho_da_amostra)

# Criando o histograma
plt.hist(dados, bins=30, density=True, alpha=0.6, color='g')

# Criando a curva de distribuição normal teórica
x = np.linspace(min(dados), max(dados), 100)
pdf = stats.norm.pdf(x, media, desvio_padrao)
plt.plot(x, pdf, 'r', linewidth=2)

# Adicionando título e rótulos
plt.title('Distribuição Normal')
plt.xlabel('Valores')
plt.ylabel('Frequência')

# Mostrando o gráfico
plt.show()

# Calculando Média, Mediana e Moda
media_amostra = np.mean(dados)
mediana_amostra = np.median(dados)
moda_amostra = stats.mode(dados)

print(f"Média: {media_amostra}")
print(f"Mediana: {mediana_amostra}")
print(f"Moda: {moda_amostra}")
```

### Análise de Média, Mediana e Moda:

* Em uma distribuição normal perfeita, a média, a mediana e a moda são iguais. No nosso exemplo prático, os valores serão muito próximos, demonstrando a tendência central dos dados.
     - **Média:** é a soma de todos os valores dividida pelo número total de valores.
     - **Mediana:** é o valor central que divide o conjunto de dados ao meio.
     - **Moda:** é o valor que aparece com mais frequência no conjunto de dados.
* Compreender essas medidas ajuda a confirmar se os dados seguem uma distribuição normal e a entender a centralidade dos dados.
